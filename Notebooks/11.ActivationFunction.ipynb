# Activation Functions
Activation functions apply a non-linear transformation and decide whether a neuron should be activated or not.

With non-linear transformations, the network can learn better and perform complex task. A network with just linear transformations is simply a linear regression.

In a network, after each layers, activation function is used.

Introduction to the following activation functions:
- Binary Step Function
    - not used in the building of neural netork, can decide whether a neuron should active or not
- Sigmoid
    - generally used in the last layer of the binary classification problem
- TanH (Hyperbolic Tangent)
    - scaled sigmoid function, good choice for hidden layers
- ReLU
    - used in hidden layers when not sure what activation function is to use
- Leaky ReLU
    - improved version of ReLu, truing to solve the vanishing gradient problem
- Softmax
    - gives probability as output, best for last layer in a network for multi class classifictions

Here assumed, <br>
**output = w*x + b**; where x = input, w = weights, b = bias <br>
**update_output = activation_function(output)**

```python
"""import necessary modules"""
import torch
import torch.nn as nn # activation functions are stored in this api
import torch.nn.functional as F # this module also have the activation functions

import numpy as np
import matplotlib.pyplot as plt
```

## Binary Step: Genarally don't use as Ativation Functon

```python
"""has only two possible decisions, plot to understand"""

bstep = lambda x: np.where(x>=0, 1, 0) # binary step

x=np.linspace(-10,10,10)
y=np.linspace(-10,10,1000)

plt.plot(y,bstep(y),'b', label='linspace(-10,10,100)')
plt.grid(linestyle='--')

plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Step Function')

plt.xticks([-4, -3, -2, -1, 0, 1, 2, 3, 4])
plt.yticks([-2, -1, 0, 1, 2])
plt.ylim(-2, 2)
plt.xlim(-4, 4)

plt.show()
#plt.savefig('step.png')
fig = plt.figure()
```

```python
"""example data for activation functions"""
z = torch.tensor([-1.0, 1.0, 2.0, 3.0])
```

## Softmax

```python
"""using activation function softmax on the demo data"""
output = torch.softmax(z, dim=0)
print("After applying softmax directly from torch: ",output)

sm = nn.Softmax(dim=0)
output = sm(z)
print("After applying sostmax from torch.nn: ",output)
```

## Sigmoid

```python
"""using activation function sigmoid on the demo data"""
output = torch.sigmoid(z)
print("Apply sigmoid from torch: ",output)

s = nn.Sigmoid()
output = s(z)
print("Apply sigmoid from torch.nn: ",output)
```

```python
"""plot the sigmoid function"""

sigmoid = lambda x: 1 / (1 + np.exp(-x)) # sigmoid function

x=np.linspace(-10,10,10)
y=np.linspace(-10,10,100)

fig = plt.figure()
plt.plot(y,sigmoid(y),'b', label='linspace(-10,10,100)')
plt.grid(linestyle='--')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Sigmoid Function')
plt.xticks([-4, -3, -2, -1, 0, 1, 2, 3, 4])
plt.yticks([-2, -1, 0, 1, 2])
plt.ylim(-2, 2)
plt.xlim(-4, 4)
plt.legend()
plt.show()
#plt.savefig('sigmoid.png')
fig = plt.figure()
```

## tanH

```python
"""using activation function tanH"""
output = torch.tanh(z)
print("Apply tanH from torch: ",output)

t = nn.Tanh()
output = t(z)
print("Apply tanH from torch.nn: ",output)
```

```python
"""plot the tanH function"""

tanh = lambda x: 2*sigmoid(2*x)-1 #sigmoid function

x=np.linspace(-10,10,10)
y=np.linspace(-10,10,100)

plt.plot(y,tanh(y),'b', label='linspace(-10,10,100)')
plt.grid(linestyle='--')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('TanH Function')
plt.xticks([-4, -3, -2, -1, 0, 1, 2, 3, 4])
plt.yticks([-4, -3, -2, -1, 0, 1, 2, 3, 4])
plt.ylim(-4, 4)
plt.xlim(-4, 4)
plt.legend()
plt.show()
#plt.savefig('tanh.png')
fig = plt.figure()
```

## ReLu

```python
"""using ReLu on demo data"""
output = torch.relu(z)
print("Apply ReLu from torch: ",output)

relu = nn.ReLU()
output = relu(z)
print("Apply ReLu from torch.nn: ",output)
```

```python
"""plot the ReLu function"""

relu = lambda x: np.where(x>=0, x, 0) # ReLu function

x=np.linspace(-10,10,10)
y=np.linspace(-10,10,1000)

plt.plot(y,relu(y),'b', label='linspace(-10,10,100)')
plt.grid(linestyle='--')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('ReLU')
plt.xticks([-4, -3, -2, -1, 0, 1, 2, 3, 4])
plt.yticks([-4, -3, -2, -1, 0, 1, 2, 3, 4])
plt.ylim(-4, 4)
plt.xlim(-4, 4)
plt.legend()
plt.show()
#plt.savefig('relu.png')
fig = plt.figure()
```

## Leaky ReLu

```python
"""using leaky relu on demo data"""
output = F.leaky_relu(z)
print("Applying leaky relu from torch.nn.functional: ",output)

lrelu = nn.LeakyReLU()
output = lrelu(z)
print("Applying leaky relu from torch.nn: ",output)
```

```python
"""plot the leaky relu function"""

leakyrelu = lambda x: np.where(x>=0, x, 0.1*x) # leaky relu function

x=np.linspace(-10,10,10)
y=np.linspace(-10,10,1000)

plt.plot(y,leakyrelu(y),'b', label='linspace(-10,10,100)')
plt.grid(linestyle='--')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Leaky ReLU')
plt.xticks([-4, -3, -2, -1, 0, 1, 2, 3, 4])
plt.yticks([-4, -3, -2, -1, 0, 1, 2, 3, 4])
plt.ylim(-4, 4)
plt.xlim(-4, 4)
plt.legend()
plt.show()
#plt.savefig('lrelu.png')
fig = plt.figure()
```

**To create a neural network:** <br>
- **Option One:** `nn.ReLU()` creates an `nn.Module` which can be added to an `nn.Sequential` model directly. <br>
- **Option Two:** `torch.relu` is just the functional API call to the relu function, so that can be added it in forward method ourself.

```python
"""option 1 (create nn modules)"""
class NeuralNet(nn.Module):
    def __init__(self, input_size, hidden_size):
        super(NeuralNet, self).__init__()
        self.linear1 = nn.Linear(input_size, hidden_size)
        self.relu = nn.ReLU()
        self.linear2 = nn.Linear(hidden_size, 1)
        self.sigmoid = nn.Sigmoid()
    
    def forward(self, z):
        out = self.linear1(z)
        out = self.relu(out)
        out = self.linear2(out)
        out = self.sigmoid(out)
        return out
```

```python
"""option 2 (using activation functions directly in forward pass)"""
class NeuralNet(nn.Module):
    def __init__(self, input_size, hidden_size):
        super(NeuralNet, self).__init__()
        self.linear1 = nn.Linear(input_size, hidden_size)
        self.linear2 = nn.Linear(hidden_size, 1)
    
    def forward(self, z):
        out = torch.relu(self.linear1(z))
        out = torch.sigmoid(self.linear2(out))
        return out
```
